```python
import pandas as pd
import re
import numpy as np
import operator
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sources/goods.csv')
df
```

```python
splited = df['cnvcGoodsName'].dropna().map(lambda row: re.split('[ ,;；，、|()（）.]', row))
splited
```

```python
inventory = {}
def insert_items(items):
    for item in items:
        inventory[item] = 1 if item not in inventory else inventory[item] + 1
splited.apply(insert_items)

sorted_inventory = sorted(inventory.items(), key=operator.itemgetter(1), reverse=True)
sorted_inventory
```

```python
labels = [key for key, _ in sorted_inventory]
sizes = [value for _, value in sorted_inventory]
colors = ['red', 'yellow', 'blue', 'green']

explode = (0.05, 0, 0, 0)

patches, l_text, p_text = plt.pie(sizes,  labels=labels,
                                       labeldistance=1.1, autopct='%2.0f%%', shadow=False,
                                       startangle=90, pctdistance=0.6)

plt.axis('equal')
plt.legend(loc='upper left', bbox_to_anchor=(-0.1, 1))
# loc: 表示legend的位置，包括'upper right','upper left','lower right','lower left'等
# bbox_to_anchor: 表示legend距离图形之间的距离，当出现图形与legend重叠时，可使用bbox_to_anchor进行调整legend的位置
# 由两个参数决定，第一个参数为legend距离左边的距离，第二个参数为距离下面的距离
plt.grid()
plt.show()
```
